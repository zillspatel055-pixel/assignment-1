```python
def final_price(actual_price,discount) :
        finalPrice = actual_price*(1-discount/100)
        return finalPrice


actual_price = int(input("Enter the actual price of product : "))
discount = int(input("Enter the discount percent on product :"))

result = final_price(actual_price,discount)
print(result)
```

```python
def get_delivery_charge(city='ahemdabad') :
    if(city == 'ahemdabad'):
        return 0
    else :
        return 50
city = str(input("Enter your city name : "))
city = city.lower()
delivery_charge = get_delivery_charge(city='ahemdabad')

print(f"Delivery charge is : {delivery_charge}")
```

```python
def  format_price(price, currency='INR') :
        if(currency == 'INR'):
             return f"Rs{price}"
        else :
             return f"${price}"
currency = str(input("Enter your currency : "))
price = int(input("Enter the price : "))
currency = currency.upper()
formatPrice = format_price(price, currency='INR')

print(f"{formatPrice}")
```

```python
def apply_coupon(price, coupon_code=None):
    if coupon_code == "ZOMATO10":
        return price * 0.9
    else:
        return price

price = int(input("Enter the price: "))
coupon_code = input("Enter your code: ")

formatPrice = apply_coupon(price, coupon_code)

print(formatPrice)
```
