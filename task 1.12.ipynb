```python
songs = ['Shape Of You', 'Blinding Lights', 'Levitating', 'Senorita']

result = map(lambda x : x.lower(), songs)

print(list(result))
```

```python
restaurant_ratings =  [4.2, 3.8, 4.5, 2.9, 3.5]

result = filter(lambda x : x > 4, restaurant_ratings)

print(list(result))
```

```python
from functools import reduce
Flipkart_shopping_cart = [499, 1299, 299, 799]
result = reduce(lambda x,y : x+y,Flipkart_shopping_cart)
print(result)

```

```python
counts = [950, 1500, 25000, 1200000]
format_followers = map(lambda x : print(f"{x/1000}K") if(x < 1000000) else print(f"{x/1000000}M"), counts)

print(format_followers)
    



```
